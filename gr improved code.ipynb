```python
#Instantiation

class Student:
    def __init__(self,student_name):
        self.student_name = student_name
        self.exam_type = set() #set
        self.exam_name = set() #set
        self.marks={} #dictionary
        self.grade={} #dictionary
        
    def get_personal_info(self):
        print("Name=\t" ,self.student_name)

    def get_total(self,exam_name):
        total = 0
        for i in self.marks.keys():
            if i == exam_name:
                for j in self.marks[i].values():
                    total += j
        return total

    def add_marks(self,exam_name,exam_type,marks):
        if exam_type not in self.exam_type:
            self.exam_type.add(exam_type)
        if exam_name not in self.exam_name:
            self.exam_name.add(exam_name)
            self.marks[exam_name] = {}
        
        self.marks[exam_name][exam_type] = marks

    def get_average_mark(self,exam_name):
        total = self.get_total(exam_name)
        average=total / len(self.exam_type)
        return average

    #In case u want to see the average i
    def print_average_mark(self,exam_name):
        total = self.get_total(exam_name)
        average=total / len(self.exam_type)
        print("the average score for ",exam_name," is:\t",average)
        
        #this calculates the student's grade

    def calculate_grade(self,exam_name):
        average = self.get_average_mark(exam_name)
        if  average >= 90:
            self.grade[exam_name] = "A"
        elif average >= 80 and average < 90:
            self.grade[exam_name] = "B"
        elif average >= 70 and average < 80:
            self.grade[exam_name] = "C+"
        elif average >= 60 and average < 70:
            self.grade[exam_name] = "C"
        elif average >= 50 and average < 60:
            self.grade[exam_name] = "D"
        elif average >= 40 and average < 50:
            self.grade[exam_name] = "E"
        else:
            self.grade[exam_name] = "F"

    def get_final_marks(self):
        self.get_personal_info()
        print("---RESULTS---")
        for i,j in self.grade.items():
            print("SUBJECT:\t",i,"\tGRADE:\t",j)


#Where execution starts

student1=Student("Maggie")
student1.add_marks("English","beginnig",23)
student1.add_marks("English","midterm",99)
student1.add_marks("English","end",90)
student1.print_average_mark("English")
student1.calculate_grade("English")
student1.add_marks("MATH","beginnig",45)
student1.add_marks("MATH","midterm",19)
student1.add_marks("MATH","end",90)
student1.print_average_mark("MATH")
student1.calculate_grade("MATH")
student1.get_final_marks()
```
